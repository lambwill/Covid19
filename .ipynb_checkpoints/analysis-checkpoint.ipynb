## Load administrative boundaries into Spatially Enabled DataFrames

url = 'https://opendata.arcgis.com/datasets/bb1440110108418ca07a69cc6d91aef1_0.zip?outSR=%7B%22latestWkid%22%3A27700%2C%22wkid%22%3A27700%7D'
filepath = 'Counties_December_2019_Boundaries_EN_BGC.zip'
shape_zip = urlretrieve(url)

```python
# create a dictionary to store the lookup table urls
shapefile_urls = {}

# Counties (December 2019) Boundaries EN BGC (20m generalised)
shapefile_urls['Counties_December_2019_Boundaries_EN_BGC'] = 'https://opendata.arcgis.com/datasets/bb1440110108418ca07a69cc6d91aef1_0.zip?outSR=%7B%22latestWkid%22%3A27700%2C%22wkid%22%3A27700%7D'

# Local Authority Districts (December 2019) Boundaries UK BGC
shapefile_urls['Local_Authority_Districts_December_2019_Boundaries_UK_BGC'] = 'https://opendata.arcgis.com/datasets/0e07a8196454415eab18c40a54dfbbef_0.zip?outSR=%7B%22latestWkid%22%3A27700%2C%22wkid%22%3A27700%7D'

```

```python
# Read the shapefiles into a dictionary if they are in the gis folder, download them first if they are not 
shapefile_dataframes = {}
for (shapefile_name, url) in shapefile_urls.items():
    # Set the filepath & name
    shapefile_folder = os.path.join(gis_folder,shapefile_name)
    shapefile_filepath = os.path.join(shapefile_folder,shapefile_name + '.shp')
    
    try:
        # Read the shapefile into a Spatially Enabled DataFrame (requires ArcGIS GeoAccessor / GeoSeriesAccessor)
        shapefile_dataframes[shapefile_name] = pd.DataFrame.spatial.from_featureclass(shapefile_filepath)
    except:
        # Download the zipped shapefile into memory
        response = urlopen(url)
        
        # Extract the contents of the zip file
        zipfile = ZipFile(BytesIO(response.read()))
        zipfile.extractall(shapefile_folder)
        
        # Read the shapefile into a Spatially Enabled DataFrame (requires ArcGIS GeoAccessor / GeoSeriesAccessor)
        shapefile_dataframes[shapefile_name] = pd.DataFrame.spatial.from_featureclass(shapefile_filepath)
```

```python
shapefile_dataframes['Counties_December_2019_Boundaries_EN_BGC'].head()
```

```python
# Get a list of the GSS codes that refer to LADs
LAD_mask = df_daily_cases['area_type']=='LAD'
list_of_LAD_GSS_codes = list(df_daily_cases[LAD_mask].index.get_level_values('GSS_CD').drop_duplicates())
```

```python
# Get a list of the GSS codes that refer to Counties
CTY_mask = ~LAD_mask
list_of_CTY_GSS_codes = list(df_daily_cases[CTY_mask].index.get_level_values('GSS_CD').drop_duplicates())
```

```python
number_of_LAD_codes = len(list_of_LAD_GSS_codes)
number_of_LAD_codes
```

```python
number_of_CTY_codes = len(list_of_CTY_GSS_codes)
number_of_CTY_codes
```

```python
number_of_GSS_codes = len(df_daily_cases['area_type'].index.get_level_values('GSS_CD').drop_duplicates())
number_of_GSS_codes
```

```python
number_of_GSS_codes == number_of_CTY_codes + number_of_LAD_codes
```

```python
# Filter for the LADs that are found in the Covid19 data, and the code, area and shape
sdf_GSS_LADs = sdf_LADs[sdf_LADs['lad19cd'].isin(list_of_LAD_GSS_codes)][['lad19cd','lad19nm','st_areasha','SHAPE']]
```

```python
# Rename the LAD code 
sdf_GSS_LADs.rename(columns={'lad19cd':'GSS_CD','lad19nm':'GSS_NM'},inplace=True)
```

```python
sdf_GSS_LADs.set_index('GSS_CD',inplace=True)
```

```python
sdf_GSS_LADs.head()
```

```python
len(sdf_GSS_LADs)
```

```python
# Filter for the CTYs that are found in the Covid19 data, and the code, area and shape
sdf_GSS_CTYs = sdf_counties[sdf_counties['cty19cd'].isin(list_of_CTY_GSS_codes)][['cty19cd','cty19nm','st_areasha','SHAPE']]
```

```python
# Rename the CTY code 
sdf_GSS_CTYs.rename(columns={'cty19cd':'GSS_CD','cty19nm':'GSS_NM'},inplace=True)
```

```python
sdf_GSS_CTYs.set_index('GSS_CD',inplace=True)
```

```python
sdf_GSS_CTYs.head()
```

```python
len(sdf_GSS_CTYs)
```

```python
sdf_GSS_boundaries = pd.concat([sdf_GSS_LADs, sdf_GSS_CTYs])
```

```python
sdf_GSS_boundaries.head()
```

```python
show_m3 = True

if show_m3:
    # Create a GIS map 
    m3 = GIS().map('United Kingdom')
    m3.zoom = 5

    # Plot the boundary data
    sdf_GSS_boundaries.spatial.plot(map_widget=m3)

    # Show the map
    m3
```

```python
sdf_GSS_boundaries.reset_index(inplace=True)
```

```python
sdf_GSS_boundaries.head()
```

```python
# Export as a Shapefile
location = os.path.join(gis_folder,'GSS_boundaries.shp')
sdf_GSS_boundaries.spatial.to_featureclass(location=location)
```
